```python
## 1. What Quarter of the year has the most Aircraft cases?

quarter_cases = air['Quarter'].value_counts().reset_index()
quarter_cases.columns = ['Quarter', 'Crash_Count']
print(quarter_cases)
```

```python
## 2. Top 10 countries that recorded the most fatalities?

exclude_list = ['Unknown', 'N/A']

top_countries = (
    air.loc[~air['Country Only'].isin(exclude_list)]
       .groupby('Country Only')['Fatalities (air)']
       .sum()
       .sort_values(ascending=False)
       .head(10)
)

print(top_countries)
```

```python
## 3. Yearly (Decade) Trend of Survival Ratio vs Death Ratio

# Calculate survival and death per decade
survival_trend = air.groupby('decade').apply(
    lambda x: pd.Series({
        'Total_Aboard': x['Aboard'].sum(),
        'Total_Fatalities': x['Fatalities (air)'].sum()
    }),
    include_groups=False
).reset_index()


survival_trend['Survival_Rate'] = (survival_trend['Total_Aboard'] - survival_trend['Total_Fatalities']) / survival_trend['Total_Aboard']
survival_trend['Death_Rate'] = survival_trend['Total_Fatalities'] / survival_trend['Total_Aboard']

print(survival_trend)
```

```python
## 4. Survival Rate vs Death Rate per Aircraft Manufacturers. Top 20

manufacturer_stats = air.groupby('Aircraft Manufacturer').agg(
    Total_Aboard=('Aboard', 'sum'),
    Total_Fatalities=('Fatalities (air)', 'sum')
).reset_index()

manufacturer_stats['Survival_Rate'] = (manufacturer_stats['Total_Aboard'] - manufacturer_stats['Total_Fatalities']) / manufacturer_stats['Total_Aboard']
manufacturer_stats['Death_Rate'] = manufacturer_stats['Total_Fatalities'] / manufacturer_stats['Total_Aboard']

top20_manufacturers = manufacturer_stats.sort_values(by='Total_Fatalities', ascending=False).head(20)
print(top20_manufacturers)
```

```python
## 7. 

top_fatal_manufacturers = air.groupby('Aircraft Manufacturer')['Fatalities (air)'].sum().sort_values(ascending=False).head(10)
print(top_fatal_manufacturers)
```

```python
## 8.

category_deaths = air.groupby('Aircraft Category')['Fatalities (air)'].sum().sort_values(ascending=False)
print(category_deaths)
```

```python
## 9. 

continent_ground_fatalities = (
    air[air['Continent'] != 'Unknown']  # Filter out Unknown
    .groupby('Continent')['Ground']
    .sum()
    .sort_values(ascending=False)
)
print(continent_ground_fatalities)
```

```python
## 10. 

ground_fatalities_decade = air.groupby('decade').agg({
    'Ground': 'sum',
    'Fatalities (air)': 'sum',
    'Aboard': 'sum'
}).reset_index()

print(ground_fatalities_decade)
```
